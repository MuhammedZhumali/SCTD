```python
import requests
import pandas as pd

# 1. Fetch the raw JSON from GitHub
url = 'https://raw.githubusercontent.com/gtarantino/TrafficData_API/master/datasource/TrafficData.json'
resp = requests.get(url)
data = resp.json()

# 2. Inspect the structure
print(type(data))        # list or dict?
# If dict:
#   print(data.keys())
# If list:
#   print(data[0])

# 3. Normalize into a DataFrame
# — If it's a list of flat records:
df_traffic = pd.DataFrame(data)

# — If it's a dict with a nested key, e.g. {'records': [...]}, then:
# df_traffic = pd.json_normalize(data['records'])

# 4. Peek at your data
print(df_traffic.shape)
df_traffic.head()
```

```python
df_traffic.columns
```

```python
data_list = []
for timestamp in df_traffic.columns:
  for borough, locations_data in df_traffic[timestamp].items():
    if pd.notna(locations_data):
      for location, traffic_data in locations_data.items():
        if isinstance(traffic_data, dict) and 'Speed' in traffic_data:
          data_list.append({
              'timestamp': timestamp,
              'Borough': traffic_data.get('Borough', borough), # Use the nested Borough if available, otherwise use the row index
              'Speed': traffic_data['Speed']
          })

df_restructured = pd.DataFrame(data_list)
display(df_restructured.head())
display(df_restructured.info())
```

```python
# 1. Re-load weather data
import pandas as pd
weather_url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv"
df_weather = pd.read_csv(weather_url, parse_dates=["Date"])

# 2. Install Prophet & import
!pip install prophet
from prophet import Prophet
import plotly.express as px

# 3. Prepare for Prophet
df_temp = df_weather.rename(columns={'Date':'ds','Temp':'y'})

# 4. Fit model
model = Prophet()
model.fit(df_temp)

# 5. Make future frame and forecast
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# 6. Plots
model.plot(forecast)
model.plot_components(forecast)

fig = px.line(
    forecast, x='ds', y=['yhat','yhat_lower','yhat_upper'],
    title='30-Day Temperature Forecast', labels={'value':'Temp','ds':'Date'}
)
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```python
from prophet.diagnostics import cross_validation, performance_metrics

# 1. Cross-validate: train on data up to each cutoff, forecast 30 days ahead
df_cv = cross_validation(
    model,
    initial="730 days",    # use first 2 years to fit
    period="90 days",    # make a new cutoff every 3 months
    horizon="30 days",    # forecast window
)

# 2. Compute metrics (MAPE, RMSE, etc.)
df_p = performance_metrics(df_cv)
print(df_p[['horizon','mape','rmse','coverage']])
```
